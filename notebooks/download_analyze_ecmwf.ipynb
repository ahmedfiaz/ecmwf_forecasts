```python
from ecmwf.opendata import Client
import metview as mv
```

```python
client = Client("ecmwf", beta=False)
parameters = ['tp']
filename = 'aifs_single_medium-rain-acc.grib'
filename

client.retrieve(
    date=0,
    time=0,
    step=18,
    stream="oper",
    type="fc",
    levtype="sfc",
    model="aifs-single",
    param=parameters,
    target=filename
)
```

```python
data = mv.read(filename)
data.describe('tp')
data.ls()
```

```python
data.to_dataset()['tp'].plot()
```
